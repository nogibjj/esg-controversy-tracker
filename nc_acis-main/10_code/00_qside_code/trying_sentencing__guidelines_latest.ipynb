```python
# Substetting for cases that have a charge code (most do)
cases_w_charges = offs_codes.loc[~offs_codes['charged_offense_code'].isna()]
cases_w_charges
```

```python
# Which ones have different sentencing codes:
# There are 9438!! That means the funnel is super steep. Lots of cases dont get convicted for what they are charged for, plea bargin? No evidence?
cases_w_sent_code = cases_w_charges.loc[~cases_w_charges['convicted_offense_code'].isna()]
cases_w_sent_code
```

```python
# Check whether the sentences are given only to the cases that have a "convicted offence code" (funnel)
cases_ready_chart = cases_w_sent_code.loc[~cases_w_sent_code['min_sentence_day'].isna()]
cases_ready_chart = cases_ready_chart.loc[~cases_ready_chart['prior_record_points'].isna()]
cases_ready_chart = cases_ready_chart[
    [
        "case_id",
        "convicted_offense_code",
        "charged_offense_date",
        "plea_code",
        "verdict_code",
        "disposition",
        "prior_record_points",
        "offense_class",
        "CL",
        "Offense Description",
        "min_sentence_day",
        "max_sentence_day",
    ]
]
cases_ready_chart
```

```python
# Subste for felonies only (we will make the misdemeanors work later LOL)
# NOPE no subsetting anymore, we got misdemeanors too
offense=cases_ready_chart
```

```python
offense['offense_type'] = 'Felony'
offense.loc[offense['offense_class'].isin(['1','2','3','A1']),'offense_type'] = 'Misdemeanor'
```

# Part 2: We try to fit into the sentencing grid boxes

```python
# code prior points
offense['prior_points_range'] = 0
for index, row in offense.iterrows():
    point = row['prior_record_points']
    if row['offense_type'] == 'felony':
        if (point >=0) & (point <=1):
            offense.loc[index,'prior_points_range'] = 1
        elif (point >=2) & (point <=5):
            offense.loc[index,'prior_points_range'] = 2
        elif (point >=6) & (point <=9):
            offense.loc[index,'prior_points_range'] = 3
        elif (point >=10) & (point <=13):
            offense.loc[index,'prior_points_range'] = 4
        elif (point >=14) & (point <=17):
            offense.loc[index,'prior_points_range'] = 5
        elif point>=18:
            offense.loc[index,'prior_points_range'] = 6
    else:
        if(point ==0):
            offense.loc[index,'prior_points_range'] = 1
        elif (point >=1) & (point <=4):
            offense.loc[index,'prior_points_range'] = 2
        elif (point >=5):
            offense.loc[index,'prior_points_range'] = 3

offense.sample(5)
```

```python
sentencing_grid = pd.read_csv('/Users/clarissaache/Documents/Capstone/qsides/00_raw_data/punishment_grid_db.csv')
sentencing_grid
```

```python
# merge cases that have complete info and sentence grid 
grid = pd.merge(offense, sentencing_grid, left_on=['offense_type','offense_class', 'prior_points_range'], right_on=['crime_type','sg_class', 'sg_prior_record_level'], validate="m:m", indicator=True)
```

```python
grid['race']
```

```python
grid['MIN_min_diff']= grid['min_sentence_day']-grid['sg_MIN_min']
grid['MIN_max_diff']= grid['min_sentence_day']-grid['sg_MIN_max']
grid
```

```python
# Fit cases into SUB boxes of the sentencing grid
grid['MIN_sentence_allocated_correctly']= 0
grid.loc[(grid['MIN_min_diff']>=0) & (grid['MIN_max_diff']<0) ,'MIN_sentence_allocated_correctly']=1
grid
```

```python
# Check proportion of cases that follow the rulesss
correct_allocations = grid.loc[grid['MIN_sentence_allocated_correctly']==1]

# remove duplicates in offense
correct_allocations = correct_allocations.groupby(['case_id', 'convicted_offense_code']).first()
correct_allocations
```

```python
felony = correct_allocations[correct_allocations['offense_type']=='Felony']
```
